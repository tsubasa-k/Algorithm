```python
from matplotlib import pyplot as plt
```

```python
name = input("file name:")
x = []
y = []

f = open(name)
for line in f.readlines():
    s = line.split(' ')
    x.append(int(s[1]))
    y.append(int(s[2]))
print(x)
    
f.close
```

```python
file = input("file name:")
f2 = open(file)
num = []
for line in f2.readlines():
    if line[0] != 'd':
        num.append(int(line))
num.append(num[0])
print(num)

n_x = [[]]
n_y = [[]]
for i in num:
    n_x[0].append(x[i-1])
    n_y[0].append(y[i-1])
plt.plot(n_x[0], n_y[0], color='r')
plt.scatter(n_x[0], n_y[0], color='b')

plt.title("ans_dt"+name[5])
```
